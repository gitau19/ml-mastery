```python
# Exercise 2: A dice game. 
# You roll a 6-sided die. Roll a 6: win $10. Anything else: lose $2. 
# Calculate the EV. Then simulate 100,000 games. Is this a positive or negative EV game?
p_six = 1/6
p_all = 5/6
d_wins = 10
d_lose = -2

ev = (p_six * d_wins) + (p_all * d_lose)
print(f"ev after each flip: ${ev:,.2f}")
print(f"ev after 100 flips: ${ev*100:,.2f}")
```

```python
d_profits = []
for roll in range(100000):
    balance = 0
    if random.random() <= p_six:
        balance += d_wins
    else:
        balance += d_lose
    d_profits.append(balance)

avg_d_profits = sum(d_profits)/len(d_profits)
print(f"Average profit over 100,000 rolls: {avg_d_profits:,.2f}")        
```
