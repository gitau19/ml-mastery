```python
def expected_values(win_prob, win_amount, lose_amount):
    lose_prob = 1 - win_prob
    return(win_prob * win_amount) + (lose_prob * lose_amount)

print(expected_values(0.52, 300, -250))
```

1. Alpha is the return ABOVE what the market gave you for free. If BTC returned 20% and your bot returned 25%, your alpha is 5%. That 5% is the value your model actually created. The other 20% was just the market doing its thing.

2. Beta is how much your returns move with the market. If BTC goes up 10% and your bot goes up 10%, your beta is 1.0 — you're just tracking the market. If your bot goes up 15%, your beta is 1.5 — you're amplifying the market. That's not skill. That's leverage.

3. Alpha = your return - (beta × market return)

4. Your bot made 30%. BTC made 20%. Sounds good. But your beta is 1.5 (you were holding leveraged positions). Expected return from beta alone: 1.5 × 20% = 30%. Your alpha: 30% - 30% = 0%. Zero. Your bot added nothing. It just took more risk.

5. Beta = covariance(bot returns, market returns) / variance(market returns)

6. Sharpe Ratio = (bot return - risk-free return) / volatility

```python
# Using your BTC data from earlier
btc["returns"] = btc["Close"].pct_change()

mean_daily_return = btc["returns"].mean()
daily_std = btc["returns"].std()
risk_free_daily = 0.05 / 365  # 5% annual, converted to daily

sharpe_daily = (mean_daily_return - risk_free_daily) / daily_std

# Annualise: multiply by sqrt(365)
sharpe_annual = sharpe_daily * (365 ** 0.5)
print(f"BTC Buy-and-Hold Sharpe: {sharpe_annual:.2f}")
```
