```python
# Exercise 1: Simulate a dice game where you win $4 on a 6, lose $1 otherwise. 
# Run 100,000 single-game trials. Then ask: is 50,000 trials enough to see the true EV? 
# Run 50k, 100k, 500k, 1M and print the average for each. Watch the average stabilise.
p_win = 1/6

win = 4
lose = -1

def simulate_roll(n_trials):
    total_profit = 0
        
    for _ in range(n_trials):
        if random.random() < p_win:
            total_profit += win
        else:
            total_profit += lose

    avg_profit = total_profit / n_trials
    return avg_profit
    
trial_sizes = [50000, 100000, 500000, 1000000]

for n in trial_sizes:
    avg = simulate_roll(n)
    print(f"{n:,} trials, average profit: {avg:,.2f}")
```

```python
# Exercise 2: European roulette has 37 slots (only one green, not two). 
# Calculate the EV for a $10 bet on red. Simulate 10,000 players × 1000 spins. 
# How does the house edge compare to American roulette? How many more players win?
def simulate_spin(num_spins, bet_size):
    balance = 0
    p_win = 18/37

    for spin in range(num_spins):
        if random.random() < p_win:
            balance += bet_size
        else:
            balance -= bet_size
    return balance

n_players = 10000
num_spins = 1000
bet_size = 10
profits = []

for player in range(n_players):
    final_balance = simulate_spin(num_spins, bet_size)
    profits.append(final_balance)

avg_results = sum(profits)/ len(profits)
winners = sum(1 for p in profits if p > 0)
losers = sum(1 for p in profits if p < 0)

print(f"Average results of 10,000 players x 1000 spins: {avg_results:,.2f}")
print(f"Winners: {winners} ({winners/len(profits)*100}%)")
print(f"Losers: {losers} ({losers/ len(profits)*100}%)")

EV = (18/37 * 10) + (19/37 * -10)
print(EV)

# house edge = expected loss/ bet size
HE = EV / 10 * 100
print(f"{HE:.2f}%")
```

```python
# A roulette wheel has 18 red, 18 black, 2 green. 
# You bet $10 on red every spin for 1,000 spins. What is your expected total loss? 
# What if you doubled your bet after every loss (Martingale) — does your expected loss change?

def simulate_spin(num_spin, bet_size):
    p_win = 18/38
    balance = 0

    for spin in range(num_spin):
        if random.random() < p_win:
            balance += bet_size
        else:
            balance -= bet_size
    return balance

results = simulate_spin(1000, 10)
print(f"1000 spins x $10 results: {results:,.2f}")


```

```python
def martingale(bankroll, base_bet, max_spins):

    balance = bankroll
    p_win = 18/38
    bet = base_bet

    for spin in range(max_spins):
        if balance < bet:
            break

        if random.random() < p_win:
            balance += bet
            bet = base_bet
        else:
            balance -= bet
            bet = bet * 2
    return balance

final = martingale(1000, 10, 1000)
print(f"Martingale bankroll: {final:,.2f}")
```

```python
flat_results = []
for _ in range(10000):
    flat_results.append(simulate_spin(1000, 10))

martingale_results = []
for _ in range(10000):
    martingale_results.append(martingale(1000, 10, 1000))

print(f"Flat betting avg: ${sum(flat_results)/ len(flat_results):,.2f}")

print(f"Martingale avg: ${sum(martingale_results)/ len(martingale_results):,.2f}")
```
