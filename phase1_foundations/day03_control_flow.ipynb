```python
# Simulate: at what N can we reliably distinguish
# a fair coin (p=0.5) from a biased coin (p=0.6)?

def fraction_heads(p_heads, num_flips):
    """Return the fraction of heads in num_flips flips."""
    heads = sum(1 for _ in range(num_flips) if random.random() < p_heads)
    return heads / num_flips

# Run 1000 experiments at different sample sizes
for n in [10, 50, 100, 500, 1000, 5000]:
    fair_results = [fraction_heads(0.5, n) for _ in range(1000)]
    biased_results = [fraction_heads(0.6, n) for _ in range(1000)]
    
    fair_avg = sum(fair_results) / len(fair_results)
    biased_avg = sum(biased_results) / len(biased_results)
    
    # Can we tell them apart? Look at the spread
    fair_min = min(fair_results)
    fair_max = max(fair_results)
    biased_min = min(biased_results)
    biased_max = max(biased_results)
    
    overlap = fair_max > biased_min  # do the ranges overlap?
    
    print(f"N={n:5d} | Fair: {fair_avg:.3f} [{fair_min:.3f}-{fair_max:.3f}] | "
          f"Biased: {biased_avg:.3f} [{biased_min:.3f}-{biased_max:.3f}] | "
          f"Overlap: {overlap}")
```

```python
import random

switch_wins = 0
stay_wins = 0

for _ in range(10000):
    car = random.randint(0, 2)
    pick = random.randint(0, 2)

    if pick == car:
        stay_wins += 1
    else:
        switch_wins += 1

print(f"Stay win rate: {stay_wins/10000:.4f}")
print(f"Switch win rate: {switch_wins/10000:.4f}")
```

```python
from math import sqrt

# Your observed win rate
win_rate = 0.60

# The "just luck" rate (null hypothesis)
p_null = 0.50

# The gap you need to prove is real
gap = win_rate - p_null  # 0.10

# 1.96 is the 95% confidence multiplier
# It means: 95% of random outcomes fall within 1.96 
# standard errors of the mean
# It comes from the normal distribution:
# - 68% of outcomes fall within 1 SE
# - 95% fall within 1.96 SE
# - 99% fall within 2.58 SE
# These are fixed numbers. You just memorise them.
z = 1.96

# Solve for n:
# gap > z * sqrt(p_null * (1 - p_null) / n)
# The full sentence reads: "Is my gap (0.10) bigger than the randomness zone at n trades?"
# If yes → your edge is real. If no → could still be luck.
# gap / z > sqrt(0.25 / n)
# (gap / z)^2 > 0.25 / n
# n > 0.25 / (gap / z)^2

n = 0.25 / (gap / z) ** 2
print(f"Trades needed for 95% confidence: {n:.0f}")

# Now see how it changes with different win rates
for wr in [0.55, 0.52, 0.51, 0.60, 0.70]:
    gap = wr - p_null
    n = 0.25 / (gap / z) ** 2
    print(f"Win rate {wr:.0%} → need {n:,.0f} trades")
```

1. Control flow — if/elif/else, logical operators, truthiness. You need this because every trading bot is a decision engine. "If signal is strong AND risk is low AND no existing position, then buy." That's control flow. Best practice: name your conditions as booleans, then branch on them.

2. Probability has two schools — Frequentist (count from data) and Bayesian (update beliefs with new evidence). Your bot will use both. Backtesting is frequentist. Live trading where you adjust to new information is Bayesian.
Sample size is everything. Same win rate can be luck or skill depending on how many trades you have. 60% over 10 trades means nothing. 60% over 96 trades is statistically significant. This matters because your bot will generate results and you need to know when to trust them.

3. P-value — the probability that randomness alone explains your result. Below 5% means your result is probably real. Above 5% means you can't rule out luck.
The formula to calculate how many trades you need:
n > 0.25 / (gap / 1.96)²
Thinner edge = exponentially more trades needed to prove it's real.

4. Why this matters for your scalper: thin edges need massive sample sizes. But scalpers generate trades fast. The real killer isn't win rate — it's whether your edge survives transaction costs.
