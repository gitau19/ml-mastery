```python
trades = [10000, 11000, 9500, 10500, 8000, 9000]

def max_drawdown(trades):
    peak = trades[0]
    max_drop = 0

    for t in trades:
        if t > peak:
            peak = t
        drop = peak - t
        if drop > max_drop:
            max_drop = drop
            
    return max_drop, max_drop/ peak * 100

drop, pct = max_drawdown(trades)
print(drop, pct)
```

```python
# Three inputs:

# 0.0005 — the centre of the bell curve (mean). This is 0.05%, representing a small average daily return.
# 0.03 — the spread (standard deviation). This is 3%, representing how wildly returns swing.
# 10000 — how many random numbers to generate.
# The difference from random.gauss: that gives you one number per call. np.random.normal gives you 10,000 numbers in one call, already in an array. No loop needed. That's the NumPy advantage.

returns = np.random.normal(0.0005, 0.03, 10000)

print(f"Mean: {np.mean(returns):.4f}")
print(f"Std: {np.std(returns):.4f}")
print(f"Min: {np.min(returns):.4f}")
print(f"Max: {np.max(returns):.4f}")
print(f"Median: {np.median(returns):.4f}")

# How many returns are beyond 2 sigma?
beyond_2sigma = np.sum(np.abs(returns) > 0.06)
print(f"Beyond 2 sigma: {beyond_2sigma}")

# Cumulative sum — turns returns into a balance curve
cumulative = np.cumsum(returns)
print(cumulative)
```

```python
returns = np.random.normal(0, 0.03, 10000)

# Get only positive returns
positive = returns[returns > 0]
print(f"Positive days: {len(positive)}")

# Get only extreme returns (beyond 2 sigma)
extreme = returns[np.abs(returns) > 0.06]
print(f"Extreme days: {len(extreme)}")

# Average of positive vs negative days
avg_win = np.mean(returns[returns > 0])
avg_loss = np.mean(returns[returns < 0])
print(f"Avg win: {avg_win:.4f}, Avg loss: {avg_loss:.4f}")
```

```python
n_players = 10000
n_spins = 1000
bet = 10
p_win = 18/38

# Generate all spins at once: True = win, False = lose
spins = np.random.random((n_players, n_spins)) < p_win

# Convert to profit: win = +bet, lose = -bet
profits = np.where(spins, bet, -bet)

# Sum each player's total
totals = np.sum(profits, axis=1)

print(f"Average result: ${np.mean(totals):,.2f}")
print(f"Winners: {np.sum(totals > 0)} ({np.mean(totals > 0)*100:.1f}%)")
```
